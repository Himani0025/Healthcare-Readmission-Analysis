```python
df = pd.read_csv("/content/Healthcare_dataset.csv")
```

```python
# Handle Missing Values
```

```python
# Replace ? with null:

df.replace("?", np.nan, inplace=True)
```

```python
df.head()
```

```python
# Check Missing Values

df.isnull().sum()
```

```python
# Drop High Missing Columns

# Weight column has too many null Values

df.drop(columns=["weight"], inplace=True)
```

```python
# Check Readmission Values

df['readmitted'].value_counts()
```

```python
# Create Simple Readmission Flag Column

df['readmitted_flag'] = df['readmitted'].apply(lambda x:0 if x == 'NO' else 1)
```

```python
df.head()
```

```python
# Calculate Readmission Rate

readmission_rate = df['readmitted_flag'].mean() * 100
print(readmission_rate)
```

```python
# Average hospital stay

df['time_in_hospital'].mean()
```

```python
# Medication vs Readmission

df.groupby('num_medications')['readmitted_flag'].mean()
```

```python
df.to_csv("clean_healthcare_data.csv", index=False)
```

```python
# Readmission Count (Bar Chart)

sns.countplot(x='readmitted_flag', data=df)
plt.title("Readmission Count")
plt.xlabel("Readmitted (0 = No, 1 = Yes)")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# Age Distribution (Histogram)

plt.hist(df["age"], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# Readmission by Age

sns.countplot(x = "age", hue="readmitted_flag", data=df)
plt.title("Readmission by Age Group")
plt.show()
```
